# Ex7-7Add2 資料視覺化教學：移動平均(Moving Average)函數的四種常用實作

本教學介紹四種常用的移動平均(Moving Average)實作方式：
1. 純Python迴圈版本
2. NumPy + convolve 捲積版（深度學習CNN常見）
3. NumPy + sliding_window_view 快速切片版
4. NumPy + cumsum 差分法（累積和）

接下來用同組資料 [1,2,3,4,5,6,7] 與 window=3 進行說明、效能比較與視覺化。

```python
# 準備基礎資料
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
a = np.array([1,2,3,4,5,6,7])
window = 3
```

## (A) 純Python迴圈版本
利用最基本的for迴圈，每次取出連續n個元素後計算平均。

```python
def moving_average_python(a, n=3):
    res = []
    for i in range(len(a)-n+1):
        res.append(sum(a[i:i+n])/n)
    return np.array(res)

print(moving_average_python(a, window))
```

## (B) NumPy + convolve 捲積法
利用np.convolve進行一維捲積，速度快且是深度學習常見技巧。

```python
def moving_average_convolve(a, n=3):
    return np.convolve(a, np.ones(n)/n, mode='valid')

print(moving_average_convolve(a, window))
```

## (C) NumPy + sliding_window_view
以sliding_window_view快速產出所有連續長度n的分段，再直接取均值。

```python
from numpy.lib.stride_tricks import sliding_window_view

def moving_average_sliding(a, n=3):
    return sliding_window_view(a, n).mean(axis=-1)

print(moving_average_sliding(a, window))
```

## (D) NumPy + cumsum 差分法
利用前綴和（累積和），差分後除以n即為移動平均。
這種方法效率極高。

```python
def moving_average_cumsum(a, n=3):
    a = np.array(a)
    cs = np.cumsum(np.insert(a, 0, 0))
    return (cs[n:] - cs[:-n]) / n

print(moving_average_cumsum(a, window))
```

## 四種方法運算結果對照
應全部相同：[2. 3. 4. 5. 6.]


```python
result_py = moving_average_python(a, window)
result_conv = moving_average_convolve(a, window)
result_slide = moving_average_sliding(a, window)
result_cumsum = moving_average_cumsum(a, window)
print('Python:', result_py)
print('Convolve:', result_conv)
print('Sliding:', result_slide)
print('Cumsum:', result_cumsum)
```

## 視覺化：原序列與移動平均
可以看出移動平均會平滑原始資料

```python
plt.figure(figsize=(8,4))
plt.plot(a, label='Original', marker='o')
plt.plot(np.arange(window-1, len(a)), result_py, label='Moving Average (window=3)', marker='s')
plt.title('Moving Average vs Original')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

## %timeit 效能比較圖（小資料）
Jupyter下執行可自動繪圖。

```python
import timeit
methods = [
    ('Python', lambda: moving_average_python(a, window)),
    ('Convolve', lambda: moving_average_convolve(a, window)),
    ('Sliding', lambda: moving_average_sliding(a, window)),
    ('Cumsum', lambda: moving_average_cumsum(a, window))
]
times = []
for name, func in methods:
    t = timeit.timeit(func, number=10000)
    times.append(t)
    print(f'{name:<10s}: {t:.6f} sec (10000次)')

# 效能柱狀圖
plt.figure(figsize=(8,4))
plt.bar([x[0] for x in methods], times, color=['r','g','b','orange'])
plt.ylabel('Total Time (seconds, 10000 runs)')
plt.title('Performance comparison (small data)')
plt.show()
```

## 大量資料效能比較（n=1000000）

```python
a_large = np.random.rand(1000000)
window_large = 100
methods_large = [
    ('Python', lambda: moving_average_python(a_large, window_large)),
    ('Convolve', lambda: moving_average_convolve(a_large, window_large)),
    ('Sliding', lambda: moving_average_sliding(a_large, window_large)),
    ('Cumsum', lambda: moving_average_cumsum(a_large, window_large))
]
times_large = []
for name, func in methods_large:
    t = timeit.timeit(func, number=10)
    times_large.append(t)
    print(f'{name:<10s}: {t:.6f} sec (10次)')

# 大量資料效能比較圖
plt.figure(figsize=(8,4))
plt.bar([x[0] for x in methods_large], times_large, color=['r','g','b','orange'])
plt.ylabel('Total Time (seconds, 10 runs)')
plt.title('Performance Comparison (Large Data)')
plt.show()
```

## 總結與建議
- 四種方法在小資料時都可用，大量資料時以cumsum最快，純Python最慢
- 建議Numpy用戶以convolve/cumsum/sliding_window為主
- sliding_window_view需Numpy 1.20+
