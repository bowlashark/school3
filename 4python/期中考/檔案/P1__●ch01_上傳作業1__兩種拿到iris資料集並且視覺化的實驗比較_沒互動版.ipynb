# ●ch01_上傳作業1_兩種拿到iris資料集並且視覺化的實驗比較_沒互動版

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt #  解決 中文字型 的問題
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']  # 設定中文為 Microsoft JhengHei
plt.rcParams['axes.unicode_minus'] = False  # 解決負號顯示問題


##  方法1： sklearn + matplotlib
iris = datasets.load_iris() # 載入 iris 資料集 (使用 sklearn 方法)
x = iris['data']
y = iris['target']
merged_array = np.c_[x, y]
df1 = pd.DataFrame(data=merged_array, columns=iris['feature_names'] + ['target'])


##  方法2： 只用 seaborn( 簡稱sns) 
df2 = sns.load_dataset("iris") # 載入 iris 資料集 (使用 seaborn 方法)


## 顯示兩方法拿到的資料集，
no_座號=99
data1=df1.iloc[no_座號:no_座號+2].to_string(line_width=120)
print(f'我的座號是{no_座號},在兩方法拿到的資料集，在index附近的資料內容是：\n', data1, "\n",df2.iloc[no_座號:no_座號+2])

# 創建畫布，使用 1 行 2 列的子圖
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# 左邊: matplotlib 視覺化
axes[0].scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], c=y, cmap='viridis')
axes[0].set_title('方法一、用sklearn取得irir並用Matplotlib視覺化_花瓣(petal)之長 vs 寬')
axes[0].set_xlabel('花瓣(petal)之長 (cm)')
axes[0].set_ylabel('花瓣(petal)之寬 (cm)')

# 右邊: seaborn 視覺化
sns.scatterplot(data=df2, x='sepal_length', y='sepal_width', hue='species', ax=axes[1])
axes[1].set_title('方法二、用Seaborn取得iris並且直接視覺化_花瓣(petal)之長 vs 寬')
axes[1].set_xlabel('花瓣(petal)之長 (cm)')
axes[1].set_ylabel('花瓣(petal)之寬 (cm)')

# 添加總標題
plt.suptitle('兩種拿到iris資料集並且視覺化的兩種方法之實驗比較', fontsize=16)
# 調整佈局並顯示圖表
plt.tight_layout()
plt.show()
```
