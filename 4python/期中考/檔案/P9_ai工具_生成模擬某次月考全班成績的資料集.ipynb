```python
###################################
# claude 生成!!!  + 一點點小小修改 
##################################
# qa:請幫忙製作一組數據，模擬某次月考全班的成績有「英文、數學、計概」三科，用python來寫，並且可放在dataframe，然後，增加修改程式，可讓第一位學生姓名改為「江俊廷」，學號是「987654321」，成績可修改為99分,98分,97分。完成後，全部放在一個JSON的檔案中。
import pandas as pd
import numpy as np
import json
from datetime import datetime

# 生成模擬數據
def generate_student_data(num_students=30):
    # 生成學號（從1開始）
    student_ids = [str(i).zfill(9) for i in range(1, num_students + 1)]
    
    # 生成姓名（使用簡單的命名方式）
    last_names = ['王', '李', '張', '劉', '陳', '林', '黃', '吳', '趙', '周']
    first_names = ['小明', '小華', '小英', '小強', '小美', '小龍', '小芳', '小玲', '小傑', '小婷']
    names = [np.random.choice(last_names) + np.random.choice(first_names) for _ in range(num_students)]
    
    # 生成各科成績（正態分佈，平均75分，標準差10）
    english_scores = np.random.normal(75, 10, num_students).round().astype(int)
    math_scores = np.random.normal(75, 10, num_students).round().astype(int)
    cs_scores = np.random.normal(75, 10, num_students).round().astype(int)
    
    # 確保成績在0-100之間
    english_scores = np.clip(english_scores, 0, 100)
    math_scores = np.clip(math_scores, 0, 100)
    cs_scores = np.clip(cs_scores, 0, 100)
    
    # 創建DataFrame
    df = pd.DataFrame({
        '學號': student_ids,
        '姓名': names,
        '英文': english_scores,
        '數學': math_scores,
        '計概': cs_scores
    })
    
    # 計算平均分數
    df['平均'] = df[['英文', '數學', '計概']].mean(axis=1).round(2)
    
    return df

# 修改特定學生資料
def modify_student_data(df, student_index=0, new_name='江俊廷', new_id='987654321', 
                       new_scores={'英文': 99, '數學': 98, '計概': 97}):
    df.loc[student_index, '姓名'] = new_name
    df.loc[student_index, '學號'] = new_id
    for subject, score in new_scores.items():
        df.loc[student_index, subject] = score
    # 重新計算平均分數
    df.loc[student_index, '平均'] = df.loc[student_index, ['英文', '數學', '計概']].mean()
    return df

# 將數據保存為JSON
def save_to_json(df, filename='student_grades.json'):
    # 將DataFrame轉換為dict，並添加額外信息
    data_dict = {
        'metadata': {
            'generated_time': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
            'total_students': len(df),
            'subjects': ['英文', '數學', '計概']
        },
        'data': df.to_dict(orient='records')
    }
    
    # 保存為JSON文件
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(data_dict, f, ensure_ascii=False, indent=2)

# 生成原始數據
df = generate_student_data()

# 修改第一位學生的資料
new_scores = {'英文': 99, '數學': 98, '計概': 97}
df = modify_student_data(df, student_index=0,new_scores=new_scores)
save_to_json(df) # 保存為JSON
# 顯示結果
print("學生成績數據__目前只顯示前面3個人的狀況。")
print(df[:3])
#print("\n成績統計：")
#print(df.describe())
print("\n數據已保存至 student_grades.json")
```

```python
import pandas as pd
import numpy as np
import json
from datetime import datetime

# 設定隨機種子確保結果可重現
np.random.seed(42)

# 生成模擬數據
def generate_student_data(num_students=30):
    # 生成學號 (從111000001開始)
    student_ids = [f"{111000001 + i}" for i in range(num_students)]
    
    # 生成姓名（用簡單的代號）
    names = [f"學生{i+1}" for i in range(num_students)]
    
    # 生成隨機成績（60-100之間）
    english_scores = np.random.randint(60, 101, num_students)
    math_scores = np.random.randint(60, 101, num_students)
    cs_scores = np.random.randint(60, 101, num_students)
    
    # 創建DataFrame
    df = pd.DataFrame({
        '學號': student_ids,
        '姓名': names,
        '英文': english_scores,
        '數學': math_scores,
        '計概': cs_scores
    })
    
    return df

# 修改特定學生資料
def modify_student_data(df):
    # 修改第一位學生的資料
    df.loc[0, '學號'] = '987654321'
    df.loc[0, '姓名'] = '江俊廷'
    df.loc[0, '英文'] = 99
    df.loc[0, '數學'] = 98
    df.loc[0, '計概'] = 97
    
    return df

# 將數據保存為JSON
def save_to_json(df, filename='student_scores.json'):
    # 將DataFrame轉換為字典，然後保存為JSON
    data_dict = df.to_dict(orient='records')
    
    # 添加元數據
    json_data = {
        'metadata': {
            'created_at': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
            'total_students': len(df),
            'subjects': ['英文', '數學', '計概']
        },
        'data': data_dict
    }
    
    # 保存為JSON文件
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(json_data, f, ensure_ascii=False, indent=2)

# 執行主程序
def main():
    # 生成原始數據
    df = generate_student_data()
    
    # 修改特定學生資料
    df = modify_student_data(df)
    
    # 保存為JSON
    save_to_json(df)
    
    # 顯示結果
    print("原始數據：")
    print(df.head())
    print("\nJSON檔案已生成：student_scores.json")

if __name__ == "__main__":
    main()
```
