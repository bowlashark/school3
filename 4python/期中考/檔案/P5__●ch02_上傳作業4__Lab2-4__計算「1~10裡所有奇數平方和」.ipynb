### 相同問題：「所有奇數平方和」，上格是JavaScript的寫法，下一格換python的寫法

```python
# 程式目的：  篩選出奇數，並將其平方，然後求和
# # pytho方法1: 用最佳的list compreshension(列表理解)來實現。
nums = range(1, 251 +1 ) 
odd_squares = [x * x for x in nums if x%2 != 0] 
print("所有奇數平方和",sum(odd_squares) )
```

```python
# pytho方法2: 使用 filter(), map(), reduce() 聚合函數來實現
from functools import reduce
nums = list(range(1,251+1 ))
result = reduce(
    lambda x, y: x + y,
    map(lambda x: x**2,
    filter(lambda x: x % 2 != 0, nums)))
print("所有奇數平方和", result)
```
