### ※JSON讀入

```python
print('dat準備資料')    
dat='----------'  ##error 測試
dat='"123","456"'  ##error 測試
dat="abcccccccccccccccccccccc"   ##error 測試
dat='{ "name江俊廷":"987654321", "year":20}'

with open('in.json', 'w' , encoding='utf-8') as f:     
    f.write(dat)


print('\ndat準備好了---用文字檔讀檔檢查檔案的內容')    
##用方法三印出內容、●● 這是配合  with 的高效率寫法！
with open('in.json',"r", encoding='utf-8') as fp:
    for line in fp:
        print(line, end='')

print('\ndat準備好了---用json_直接載入load')    
###  JSON讀入######################
dat_load=''
import json
with open('in.json',"r", encoding='utf-8') as fp:
    dat_load=json.load(fp)
print(dat_load) ##直接是正確的 json的物件格式！
```
