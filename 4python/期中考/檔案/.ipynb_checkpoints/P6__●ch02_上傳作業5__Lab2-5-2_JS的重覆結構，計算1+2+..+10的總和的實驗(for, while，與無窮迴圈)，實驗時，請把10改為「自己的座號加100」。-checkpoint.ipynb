### 相同問題：「1+2+..+10的總和」，上格是JavaScript的寫法，下一格換python的寫法

```python
#方法1
t=0
for i in range(351+1 ):
    t+=i
print("方法1", t)

#方法2
t=0
i=1
while i<=351:    
    t+=i
    i+=1
print("方法2", t)

#方法3
t=0
i=1
while True:
    if i>351:
        break
    t+=i
    i+=1
print("方法3", t)
```
