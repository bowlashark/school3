```python
# å°å…¥å¿…è¦çš„å¥—ä»¶
import csv
import json
import sqlite3
import pandas as pd
from pymongo import MongoClient
import os
```

## åˆå§‹å­¸ç”Ÿè³‡æ–™

```python
# å®šç¾©åˆå§‹å­¸ç”Ÿè³‡æ–™
students_data = [
    {"å­¸è™Ÿ": "S001", "å§“å": "å¼µå°æ˜", "ä½å€": "å°åŒ—å¸‚ä¿¡ç¾©å€", "é›»è©±": "02-12345678"},
    {"å­¸è™Ÿ": "S002", "å§“å": "æç¾éº—", "ä½å€": "æ–°åŒ—å¸‚æ¿æ©‹å€", "é›»è©±": "02-23456789"},
    {"å­¸è™Ÿ": "S003", "å§“å": "ç‹å¤§åŒ", "ä½å€": "å°ä¸­å¸‚è¥¿å±¯å€", "é›»è©±": "04-34567890"},
    {"å­¸è™Ÿ": "S004", "å§“å": "é™³å°ç‰", "ä½å€": "é«˜é›„å¸‚å·¦ç‡Ÿå€", "é›»è©±": "07-45678901"}
]

print("åˆå§‹å­¸ç”Ÿè³‡æ–™ï¼š")
for student in students_data:
    print(student)
```

# 1. CSV æª”æ¡ˆå„²å­˜æ–¹å¼

```python
# CSV æª”æ¡ˆæ“ä½œ - ç°¡å–®çš„æ–‡å­—æ ¼å¼ï¼Œé©åˆè¡¨æ ¼è³‡æ–™
csv_filename = "students.csv"

# Create - å»ºç«‹CSVæª”æ¡ˆä¸¦å¯«å…¥è³‡æ–™
with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['å­¸è™Ÿ', 'å§“å', 'ä½å€', 'é›»è©±']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for student in students_data:
        writer.writerow(student)

print(f"âœ… CSVæª”æ¡ˆ {csv_filename} å»ºç«‹å®Œæˆï¼")

# Read - è®€å–CSVæª”æ¡ˆ
print("\nğŸ“– å¾CSVè®€å–çš„è³‡æ–™ï¼š")
df_csv = pd.read_csv(csv_filename)
print(df_csv)

# Update - æ›´æ–°è³‡æ–™ï¼ˆæ–°å¢ä¸€ç­†ï¼‰
new_student = {"å­¸è™Ÿ": "S005", "å§“å": "æ—å°è¯", "ä½å€": "å°å—å¸‚æ±å€", "é›»è©±": "06-56789012"}
with open(csv_filename, 'a', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writerow(new_student)

print(f"\nâœ… å·²æ–°å¢å­¸ç”Ÿ {new_student['å§“å']} åˆ°CSVæª”æ¡ˆ")

# é‡æ–°è®€å–é¡¯ç¤ºæ›´æ–°å¾Œçš„è³‡æ–™
df_csv_updated = pd.read_csv(csv_filename)
print("\nğŸ“– æ›´æ–°å¾Œçš„CSVè³‡æ–™ï¼š")
print(df_csv_updated)
```

# 2. JSON æª”æ¡ˆå„²å­˜æ–¹å¼

```python
# JSON æª”æ¡ˆæ“ä½œ - çµæ§‹åŒ–è³‡æ–™æ ¼å¼ï¼Œé©åˆè¤‡é›œè³‡æ–™çµæ§‹
json_filename = "students.json"

# Create - å»ºç«‹JSONæª”æ¡ˆ
with open(json_filename, 'w', encoding='utf-8') as jsonfile:
    json.dump(students_data, jsonfile, ensure_ascii=False, indent=2)

print(f"âœ… JSONæª”æ¡ˆ {json_filename} å»ºç«‹å®Œæˆï¼")

# Read - è®€å–JSONæª”æ¡ˆ
print("\nğŸ“– å¾JSONè®€å–çš„è³‡æ–™ï¼š")
with open(json_filename, 'r', encoding='utf-8') as jsonfile:
    json_data = json.load(jsonfile)
    for student in json_data:
        print(student)

# Update - æ›´æ–°è³‡æ–™
json_data.append(new_student)
with open(json_filename, 'w', encoding='utf-8') as jsonfile:
    json.dump(json_data, jsonfile, ensure_ascii=False, indent=2)

print(f"\nâœ… å·²æ–°å¢å­¸ç”Ÿ {new_student['å§“å']} åˆ°JSONæª”æ¡ˆ")

# Delete - åˆªé™¤è³‡æ–™ï¼ˆåˆªé™¤æœ€å¾Œä¸€ç­†ï¼‰
if len(json_data) > 0:
    deleted_student = json_data.pop()
    with open(json_filename, 'w', encoding='utf-8') as jsonfile:
        json.dump(json_data, jsonfile, ensure_ascii=False, indent=2)
    print(f"\nâŒ å·²åˆªé™¤å­¸ç”Ÿ {deleted_student['å§“å']} å¾JSONæª”æ¡ˆ")

print("\nğŸ“– æœ€çµ‚JSONè³‡æ–™ï¼š")
with open(json_filename, 'r', encoding='utf-8') as jsonfile:
    final_json_data = json.load(jsonfile)
    for student in final_json_data:
        print(student)
```

# 3. SQLite3 è³‡æ–™åº«å„²å­˜æ–¹å¼

```python
# SQLite3 è³‡æ–™åº«æ“ä½œ - è¼•é‡ç´šé—œè¯å¼è³‡æ–™åº«
db_filename = "students.db"

# é€£æ¥è³‡æ–™åº«
conn = sqlite3.connect(db_filename)
cursor = conn.cursor()

# Create - å»ºç«‹è³‡æ–™è¡¨
cursor.execute('''
    CREATE TABLE IF NOT EXISTS students (
        å­¸è™Ÿ TEXT PRIMARY KEY,
        å§“å TEXT NOT NULL,
        ä½å€ TEXT,
        é›»è©± TEXT
    )
''')

# Insert - æ’å…¥åˆå§‹è³‡æ–™
for student in students_data:
    cursor.execute('''
        INSERT OR REPLACE INTO students (å­¸è™Ÿ, å§“å, ä½å€, é›»è©±)
        VALUES (?, ?, ?, ?)
    ''', (student['å­¸è™Ÿ'], student['å§“å'], student['ä½å€'], student['é›»è©±']))

conn.commit()
print(f"âœ… SQLite3è³‡æ–™åº« {db_filename} å»ºç«‹å®Œæˆï¼")

# Read - æŸ¥è©¢è³‡æ–™
print("\nğŸ“– å¾SQLite3è®€å–çš„è³‡æ–™ï¼š")
cursor.execute("SELECT * FROM students")
rows = cursor.fetchall()
for row in rows:
    print(f"å­¸è™Ÿ: {row[0]}, å§“å: {row[1]}, ä½å€: {row[2]}, é›»è©±: {row[3]}")

# Update - æ›´æ–°è³‡æ–™
cursor.execute('''
    UPDATE students SET é›»è©± = ? WHERE å­¸è™Ÿ = ?
''', ("02-88888888", "S001"))
conn.commit()
print("\nâœ… å·²æ›´æ–°å­¸ç”Ÿ S001 çš„é›»è©±")

# Delete - åˆªé™¤è³‡æ–™
cursor.execute('''
    DELETE FROM students WHERE å­¸è™Ÿ = ?
''', ("S004",))
conn.commit()
print("âœ… å·²åˆªé™¤å­¸ç”Ÿ S004")

# æœ€çµ‚æŸ¥è©¢
print("\nğŸ“– æœ€çµ‚SQLite3è³‡æ–™ï¼š")
cursor.execute("SELECT * FROM students")
final_rows = cursor.fetchall()
for row in final_rows:
    print(f"å­¸è™Ÿ: {row[0]}, å§“å: {row[1]}, ä½å€: {row[2]}, é›»è©±: {row[3]}")

# é—œé–‰é€£æ¥
conn.close()
```

# 4. MongoDB è³‡æ–™åº«å„²å­˜æ–¹å¼ï¼ˆé¡å¤–åŠ åˆ†é¡Œï¼‰

```python
# MongoDB è³‡æ–™åº«æ“ä½œ - NoSQLæ–‡ä»¶è³‡æ–™åº«
try:
    # é€£æ¥MongoDB
    client = MongoClient('localhost', 27017)
    
    # æª¢æŸ¥é€£æ¥æ˜¯å¦æˆåŠŸ
    client.admin.command('ismaster')
    
    # é¸æ“‡è³‡æ–™åº«å’Œé›†åˆ
    db = client['school_db']
    collection = db['students']
    
    # æ¸…ç©ºç¾æœ‰è³‡æ–™ï¼ˆç‚ºäº†æ¼”ç¤ºï¼‰
    collection.delete_many({})
    
    # Create - æ’å…¥è³‡æ–™
    result = collection.insert_many(students_data)
    print(f"âœ… MongoDB è³‡æ–™æ’å…¥æˆåŠŸï¼æ’å…¥ {len(result.inserted_ids)} ç­†è³‡æ–™")
    
    # Read - æŸ¥è©¢æ‰€æœ‰è³‡æ–™
    print("\nğŸ“– å¾MongoDBè®€å–çš„è³‡æ–™ï¼š")
    all_students = collection.find()
    for student in all_students:
        print(student)
    
    # Update - æ›´æ–°è³‡æ–™
    collection.update_one(
        {"å­¸è™Ÿ": "S002"},
        {"$set": {"ä½å€": "æ–°åŒ—å¸‚æ–°åº—å€", "é›»è©±": "02-99999999"}}
    )
    print("\nâœ… å·²æ›´æ–°å­¸ç”Ÿ S002 çš„è³‡æ–™")
    
    # æŸ¥è©¢ç‰¹å®šè³‡æ–™
    print("\nğŸ” æŸ¥è©¢å­¸è™Ÿ S002 çš„è³‡æ–™ï¼š")
    student_s002 = collection.find_one({"å­¸è™Ÿ": "S002"})
    print(student_s002)
    
    # Delete - åˆªé™¤è³‡æ–™
    delete_result = collection.delete_one({"å­¸è™Ÿ": "S003"})
    print(f"\nâœ… å·²åˆªé™¤ {delete_result.deleted_count} ç­†å­¸ç”Ÿè³‡æ–™")
    
    # æœ€çµ‚æŸ¥è©¢
    print("\nğŸ“– æœ€çµ‚MongoDBè³‡æ–™ï¼š")
    final_students = collection.find()
    for student in final_students:
        print(student)
    
    # é—œé–‰é€£æ¥
    client.close()
    
except Exception as e:
    print(f"âŒ MongoDB é€£æ¥å¤±æ•—: {e}")
    print("è«‹ç¢ºèª MongoDB æœå‹™æ­£åœ¨é‹è¡Œåœ¨ port 27017")
    print("å®‰è£æŒ‡ä»¤åƒè€ƒ: brew install mongodb-community æˆ–åƒè€ƒå®˜æ–¹æ–‡ä»¶")
```

# è³‡æ–™è¦–è¦ºåŒ–æ¯”è¼ƒ

```python
# æ¯”è¼ƒå››ç¨®å„²å­˜æ–¹å¼çš„ç‰¹æ€§
storage_types = ['CSV', 'JSON', 'SQLite3', 'MongoDB']
complexity = [1, 2, 3, 4]  # è¤‡é›œåº¦è©•åˆ†
flexibility = [2, 4, 3, 5]  # éˆæ´»æ€§è©•åˆ†
ease_of_use = [5, 4, 3, 3]  # æ˜“ç”¨æ€§è©•åˆ†

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# é•·æ¢åœ–æ¯”è¼ƒ
x_pos = range(len(storage_types))
width = 0.25

axes[0].bar([x - width for x in x_pos], complexity, width, label='è¤‡é›œåº¦', alpha=0.8)
axes[0].bar(x_pos, flexibility, width, label='éˆæ´»æ€§', alpha=0.8)
axes[0].bar([x + width for x in x_pos], ease_of_use, width, label='æ˜“ç”¨æ€§', alpha=0.8)

axes[0].set_xlabel('å„²å­˜æ–¹å¼')
axes[0].set_ylabel('è©•åˆ† (1-5)')
axes[0].set_title('å››ç¨®å„²å­˜æ–¹å¼ç‰¹æ€§æ¯”è¼ƒ')
axes[0].set_xticks(x_pos)
axes[0].set_xticklabels(storage_types)
axes[0].legend()

# é›·é”åœ–æ¯”è¼ƒ
categories = ['è¤‡é›œåº¦', 'éˆæ´»æ€§', 'æ˜“ç”¨æ€§', 'æŸ¥è©¢æ•ˆèƒ½', 'æ“´å±•æ€§']
csv_stats = [1, 2, 5, 3, 2]
json_stats = [2, 4, 4, 3, 3]
sqlite_stats = [3, 3, 3, 4, 3]
mongo_stats = [4, 5, 3, 4, 5]

# é–‰åˆé›·é”åœ–
csv_stats += csv_stats[:1]
json_stats += json_stats[:1]
sqlite_stats += sqlite_stats[:1]
mongo_stats += mongo_stats[:1]
categories += categories[:1]

angles = [n / float(len(categories)-1) * 2 * 3.14159 for n in range(len(categories))]

axes[1].plot(angles, csv_stats, 'o-', linewidth=2, label='CSV')
axes[1].plot(angles, json_stats, 'o-', linewidth=2, label='JSON')
axes[1].plot(angles, sqlite_stats, 'o-', linewidth=2, label='SQLite3')
axes[1].plot(angles, mongo_stats, 'o-', linewidth=2, label='MongoDB')
axes[1].fill(angles, csv_stats, alpha=0.25)
axes[1].fill(angles, json_stats, alpha=0.25)
axes[1].fill(angles, sqlite_stats, alpha=0.25)
axes[1].fill(angles, mongo_stats, alpha=0.25)

axes[1].set_xticks(angles[:-1])
axes[1].set_xticklabels(categories[:-1])
axes[1].set_title('å„²å­˜æ–¹å¼ç¶œåˆæ¯”è¼ƒé›·é”åœ–')
axes[1].legend(loc='upper right')

plt.tight_layout()
plt.show()

print("\nğŸ“Š å››ç¨®å„²å­˜æ–¹å¼ç¸½çµï¼š")
print("1. CSV    - ç°¡å–®è¡¨æ ¼ï¼Œé©åˆåŒ¯å…¥åŒ¯å‡º")
print("2. JSON   - çµæ§‹åŒ–è³‡æ–™ï¼Œé©åˆAPIå‚³è¼¸")
print("3. SQLite3 - è¼•é‡ç´šé—œè¯å¼è³‡æ–™åº«ï¼ŒåŠŸèƒ½å®Œæ•´")
print("4. MongoDB - NoSQLæ–‡ä»¶è³‡æ–™åº«ï¼Œæ“´å±•æ€§å¼·")
```
