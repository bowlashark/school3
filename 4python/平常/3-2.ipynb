```python
import requests

# JSON
http='https://quality.data.gov.tw/dq_download_json.php?nid=173922&md5_url=af41415ea3dce7debef7e7fc79fade1a'
response = requests.get(http)
dic=response.json()
for i in range(14):   
    print(dic[i]['district'], " 地段：" , dic[i]['rps02'][:16], " 平方公尺=",dic[i]['rps03_area'], " 總價=", dic[i]['rps22_amountsunitdollars'], "  每平公尺價格=" , f"{dic[i]['rps23_amountsunitdollars']}")
```

```python
import requests

# 下載 CSV
# csv方法一：直接爬下csv數據，就當作文字檔處理
import requests

http = 'https://quality.data.gov.tw/dq_download_csv.php?nid=173922&md5_url=af41415ea3dce7debef7e7fc79fade1a'
response = requests.get(http)
dicts = response.text
dict = dicts.split('\n')
data_first_lint = dict[0].split(',')

for i in range(1, 14):
    datas = dict[i].split(',')
    for j in range(len(datas)):
        if data_first_lint[j] == 'district':
            print(datas[j] + ',', end="")
        if data_first_lint[j] == 'rps02':
            print("  地段" + datas[j] + ',', end="")
        if data_first_lint[j] == 'rps03_area':
            print("  平方公尺" + datas[j] + ',', end="")
    print()
```

```python
# csv方法二：直接爬下csv數據，先存為csv檔，再用csv.DictReader(csvfile)轉為dict處理
import requests

http = 'https://quality.data.gov.tw/dq_download_csv.php?nid=173922&md5_url=af41415ea3dce7debef7e7fc79fade1a'
response = requests.get(http)
dicts = response.text

# 先存成檔案
with open('in.csv', 'w', encoding='utf-8') as f:
    f.write(dicts)

print('\ndat準備好了---用文字檔讀檔檢查檔案的內容')
print('-' * 60)

# 用 with 逐行讀出，印前5行
with open('in.csv', "r", encoding='utf-8') as fp:
    for i, row in enumerate(fp):
        if i < 15:
            print(row, end='')
```
